```python
# Merge the DataFrames on 'Loan ID'/'Loan Id'
merged_df = pd.merge(disbursment_df, emi_repay_df, left_on='LOAN ID', right_on='Loan Id', how='inner')

```

```python
# Check for duplicate loan IDs in merged_df
duplicate_count = merged_df['Loan Id'].duplicated().sum()

# Print the count of duplicate loan IDs
print("Number of duplicate Loan IDs:", duplicate_count)
```

```python
# Get unique values in the 'Status' column
unique_statuses = merged_df['Status'].unique()

# Print the unique statuses
print("Unique Statuses:", unique_statuses)
```

```python
# Replace 'POSTPAID' with 'Delay' and 'UNPAID' with 'Default' in the 'Status' column
#merged_df['Status'].replace({'POSTPAID': 'Delay', 'UNPAID': 'Default'}, inplace=True)

```

```python
merged_df['Status'].replace({'ONTIME': 'Ontime', 'PREPAID' : 'Prepaid', 'DELAY': 'Delay','DEFAULT': 'Default' }, inplace=True)
```

```python
merged_df.head(5)
```

```python
merged_df.to_csv('complete.csv')
```

```python
def func(A):
    b = 'prepaid'
    for i in A:
        if i == 'Default':
            b = 'Default'
        if b != 'Default' and i == 'Delay':
            b = 'Delay'
        if b != 'Default' and i != 'Delay' and i == 'Ontime':
            b = 'Ontime'
    return b

# Group by 'LOAN ID' and apply the function to each group
result = merged_df.groupby('LOAN ID')['Status'].apply(func).reset_index()

# Rename the columns in the result DataFrame
result.columns = ['LOAN ID', 'Repayment Status']

# Display the result
result

```

```python
# Check for duplicate loan IDs in merged_df
duplicate_count = result['LOAN ID'].duplicated().sum()

# Print the count of duplicate loan IDs
print("Number of duplicate Loan IDs:", duplicate_count)
```
